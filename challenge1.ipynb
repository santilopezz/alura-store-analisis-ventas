### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 4. Productos más y menos vendidos

```python
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}
```

```python
for nombre_tienda, tienda in tiendas.items():
    print(f"Análisis de la tienda: {nombre_tienda}")

    productos_tienda = tienda['Producto'].value_counts()

    productos_mas_vendidos = productos_tienda.head(5)
    print("\nProductos más vendidos:")
    print(productos_mas_vendidos)

    productos_menos_vendidos = productos_tienda.tail(5)
    print("\nProductos menos vendidos:")
    print(productos_menos_vendidos)

    print("\n" + "-"*40 + "\n")
```

# Gráficos

```python
ingresos = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Facturacion': [total_tienda1, total_tienda2, total_tienda3, total_tienda4]
}
```

```python
calificaciones = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación': [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
}
```

```python
df_grafico1 = pd.DataFrame(ingresos)
```

```python
df_grafico1.plot(x='Tienda', y='Facturacion', kind='bar', color='green')
plt.title('\nComparativa de Facturación entre Tiendas\n')
plt.xlabel('Tiendas')
plt.ylabel('\nIngreso Total (Miles de millones de $)\n')
plt.grid(True, linestyle='--', alpha=0.7)
```

```python
df_grafico2.plot(x='Tienda', y='Calificación', kind='barh', color='orange')
plt.title('\nComparativa de calificaciones entre tiendas\n')
plt.xlabel('\nCalificaciones\n')
plt.grid(True, linestyle='--', alpha=0.7)
```

```python
df_grafico3.plot(x='Tienda', y='Envío', kind='line',color='red')
plt.title('\nComparativa de costo promedio de envío entre tiendas\n')
plt.xlabel('\nCosto de envío promedio\n')
plt.grid(True, linestyle='--', alpha=0.7)
```
